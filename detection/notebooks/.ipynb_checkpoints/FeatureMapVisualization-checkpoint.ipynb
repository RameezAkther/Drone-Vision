# Feature Map Visualization

## Base RT-DETR model

```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from torchvision import transforms

from rtdetr_pytorch.config import get_config  # repo’s config loader
from rtdetr_pytorch.modeling import build_model
```

```python
# --------------------------------------------------------------
# 1. Load trained RT-DETR model
# --------------------------------------------------------------
cfg = get_config("configs/rtdetr/rtdetr_r50vd_6x_coco.yml")
model = build_model(cfg, is_train=False)

checkpoint = torch.load("weights/rtdetr_trained.pth", map_location="cpu")
model.load_state_dict(checkpoint["model"])
model.eval()

backbone = model.backbone  # get backbone module
```

```python
# --------------------------------------------------------------
# 2. Image preprocessing
# --------------------------------------------------------------
def preprocess(image_path):
    img = Image.open(image_path).convert("RGB")
    tf = transforms.Compose([
        transforms.Resize((640, 640)),
        transforms.ToTensor()
    ])
    return tf(img).unsqueeze(0)
```

```python
# --------------------------------------------------------------
# 3. Register hooks for intermediate outputs
# RT-DETR backbone layers:
#   stage_1 → low level
#   stage_2 → C3
#   stage_3 → C4
#   stage_4 → C5
# --------------------------------------------------------------
feature_maps = {}

def make_hook(name):
    def hook(module, input, output):
        feature_maps[name] = output
    return hook


backbone.stage_2.register_forward_hook(make_hook("C3"))
backbone.stage_3.register_forward_hook(make_hook("C4"))
backbone.stage_4.register_forward_hook(make_hook("C5"))
```

```python
# --------------------------------------------------------------
# 4. Forward pass
# --------------------------------------------------------------
img = preprocess("./sample_inferencing_data/2.jpg")

with torch.no_grad():
    _ = model(img)  # run full RT-DETR model
```

```python
# --------------------------------------------------------------
# 5. Visualization
# --------------------------------------------------------------
def show_feature_maps(feats):
    layers = ["C3", "C4", "C5"]

    plt.figure(figsize=(15, 5))
    for i, l in enumerate(layers):
        fmap = feats[l][0].mean(0).cpu().numpy()

        plt.subplot(1, 3, i + 1)
        plt.imshow(fmap, cmap="viridis")
        plt.title(f"{l} shape={tuple(feats[l].shape)}")
        plt.axis("off")

    plt.tight_layout()
    plt.show()
```

```python
show_feature_maps(feature_maps)
```

## Redefined Backbone's Feature Map

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image

from rtdetr_pytorch.config import get_config
from rtdetr_pytorch.modeling import build_model
```

```python
def load_enhanced_rtdetr(cfg_path, weights_path):
    cfg = get_config(cfg_path)
    model = build_model(cfg, is_train=False)

    ckpt = torch.load(weights_path, map_location="cpu")
    model.load_state_dict(ckpt["model"])
    model.eval()

    return model
```

```python
def preprocess_image(path):
    img = Image.open(path).convert("RGB")
    tf = transforms.Compose([
        transforms.Resize((640, 640)),
        transforms.ToTensor(),
    ])
    return tf(img).unsqueeze(0)
```

```python
feature_maps = {
    "C3": None,
    "C4": None,
    "C5": None,
    "P3": None,
    "P4": None,
    "P5": None,
    "FSG": None
}
```

```python
def hook_backbone(stage_name):
    def hook(module, input, output):
        feature_maps[stage_name] = output
    return hook
```

```python
def hook_bifpn(stage_name):
    def hook(module, input, output):
        feature_maps[stage_name] = output
    return hook
```

```python
def hook_fsg(module, input, output):
    feature_maps["FSG"] = output[0]   # (B,1,H,W)
```

```python
def attach_hooks(model):
    # Backbone stages
    model.backbone.stage_2.register_forward_hook(hook_backbone("C3"))
    model.backbone.stage_3.register_forward_hook(hook_backbone("C4"))
    model.backbone.stage_4.register_forward_hook(hook_backbone("C5"))

    # BiFPN stages (P3, P4, P5)
    bifpn = model.neck
    bifpn.fpn_convs[0].register_forward_hook(hook_bifpn("P3"))
    bifpn.fpn_convs[1].register_forward_hook(hook_bifpn("P4"))
    bifpn.fpn_convs[2].register_forward_hook(hook_bifpn("P5"))

    # FSG (Feature Selective Gate) attention mask
    model.fusion.gate.register_forward_hook(hook_fsg)
```

```python
def run_and_collect(model, img_tensor):
    with torch.no_grad():
        _ = model(img_tensor)
```

```python
def visualize_feature_maps_real(feature_maps):
    layers = ["P2", "P3", "P4", "P5"]
    
    # Convert backbone C3/C4/C5 to P2/P3/P4/P5 resolutions
    mapping = {
        "P2": "C3",
        "P3": "P3",
        "P4": "P4",
        "P5": "P5"
    }

    fig, axes = plt.subplots(1, 4, figsize=(22, 5))

    for i, p in enumerate(layers):
        fmap = feature_maps[mapping[p]][0]   # (C,H,W)
        fmap = fmap.mean(0).cpu().numpy()

        axes[i].imshow(fmap, cmap='viridis')
        axes[i].set_title(f"{p} Feature Map\nShape={fmap.shape}")
        axes[i].axis("off")

    plt.show()
```

```python
def extract_and_visualize(cfg, weights, image_path):
    model = load_enhanced_rtdetr(cfg, weights)
    attach_hooks(model)

    img = preprocess_image(image_path)
    run_and_collect(model, img)

    visualize_feature_maps_real(feature_maps)
```

```python
extract_and_visualize(
    cfg="configs/rtdetr/enhanced_rtdetr_convnext_fsg.yml",
    weights="weights/enhanced_rtdetr_trained.pth",
    image_path="./sample_inferencing_data/2.jpg"
)
```
