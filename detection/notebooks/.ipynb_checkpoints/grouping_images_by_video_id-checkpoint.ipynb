```python
import os
import json
import shutil
from tqdm import tqdm

# Paths
json_path_train = "../dataset/seadronessee/annotations/instances_val_objects_in_fixed.json"
image_dir = r"D:\Dataset\images\val"   # adjust path to actual image folder
output_dir = r"F:\videos_grouped_val"

os.makedirs(output_dir, exist_ok=True)

# Load JSON
with open(json_path_train, 'r') as f:
    train_data = json.load(f)

# Map images to video_id
video_to_files = {}
for img in train_data["images"]:
    video_id = img["video_id"]
    file_name = img["file_name"]
    
    if video_id not in video_to_files:
        video_to_files[video_id] = []
    video_to_files[video_id].append(file_name)

# Copy images into video-wise folders
for video_id, files in tqdm(video_to_files.items(), desc="Grouping by video"):
    video_folder = os.path.join(output_dir, f"video_{video_id}")
    os.makedirs(video_folder, exist_ok=True)
    
    for fname in files:
        src = os.path.join(image_dir, fname)
        dst = os.path.join(video_folder, fname)
        
        if os.path.exists(src):
            shutil.copy(src, dst)
```
